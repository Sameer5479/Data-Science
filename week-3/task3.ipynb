```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
data=pd.read_csv('/Users/sameerkumar/Downloads/householdtask3.csv')
```

```python
display(data.head(10))
```

```python
#scatterplot
plt.scatter(data['year'],data['own'])

#Adding the title
plt.title("Scatter plot")

#Setting x and y labels
plt.xlabel('year')
plt.ylabel('own')

#adding the legends
plt.show()
```

```python
#Line chart 
plt.plot(data['year'])
plt.plot(data['own'])


#Adding the title
plt.title("Scatter plot")

#Setting x and y labels
plt.xlabel('year')
plt.ylabel('own')

#adding the legends
plt.show()
```

```python
#Bar chart or bar plot
plt.bar(data['year'],data['own'])

#Adding the title
plt.title("Scatter plot")

#Setting x and y labels
plt.xlabel('year')
plt.ylabel('own')

#adding the legends
plt.show()
```

```python
#Histogram
plt.hist(data['income'])

plt.title("Histogram")

plt.show()
```
